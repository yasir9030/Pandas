```python
import pandas as pd
import numpy as np
```

```python
data=np.random.rand(10,3)
for col in range(data.shape[1]):
  null_indicates=np.random.choice(10,3,replace=False)
  data[null_indicates,col]=np.nan
  df=pd.DataFrame(data,columns=['col1','col2','col3'])
```
